```python
Matplotlib is a popular Python library for creating static, animated, and interactive visualizations in 2D and 3D. It provides a flexible and comprehensive framework for creating a wide range of plots and charts. Matplotlib is widely used for data visualization in scientific research, data analysis, and various fields of data science and engineering.

=>Matplotlib allows users to create a wide variety of plots and charts to visualize data effectively, making it easier to understand patterns, trends, and relationships within the data.

Typer of plots created in matplotlib:
    
Line Plot: Line plots are used to display data points connected by lines, making them suitable for showing trends or changes over time.

Scatter Plot: Scatter plots are used to display individual data points as dots on a graph, making them useful for visualizing relationships between two variables.

Bar Plot: Bar plots are used to represent categorical data with rectangular bars, making them suitable for comparing values of different categories.

Histogram: Histograms are used to visualize the distribution of a single variable by dividing the data into bins and displaying the frequency or density of data points in each bin.

Pie Chart: Pie charts represent data as slices of a circle, with each slice corresponding to a category or proportion of the whole. They are used to visualize the composition or distribution of a dataset.

Matplotlib's flexibility allows users to create many other types of plots, including box plots, violin plots, 3D plots, error bars, and more, making it a powerful tool for data visualization in Python.





```

```python
#A scatter plot is a type of data visualization that is used to display individual data points on a two-dimensional Cartesian plane. In a scatter plot, each data point is represented as a single dot or marker on the graph, with its position determined by the values of two variables. The horizontal axis (x-axis) typically represents one variable, and the vertical axis (y-axis) represents the other variable. Scatter plots are useful for visualizing the relationship or correlation between these two variables.
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.xlabel("x -axis",)
plt.ylabel("y-axis")
plt.title("this is title")
plt.scatter(x,y,color="red")
```

```python
#The subplot() function in Matplotlib is used to create a grid of subplots within a single figure, allowing you to visualize multiple plots in an organized manner. 
import numpy as np
import matplotlib.pyplot as plt
plt.subplot(2, 2, 1) 
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
plt.plot(x1, y1, marker='o',color='b')
plt.title('Line Plot 1')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.subplot(2,2, 2)  
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
plt.plot(x2, y2, marker='s',color='g')
plt.title('Line Plot 2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.subplot(2, 2, 3)  
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
plt.plot(x3, y3, marker='^', linestyle='-', color='r')
plt.title('Line Plot 3')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.subplot(2, 2, 4)  
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.plot(x4, y4,marker="s",linestyle=':', color='purple')
plt.title('Line Plot 4')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

```

```python

import numpy as np
import matplotlib.pyplot as plt
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1, box2]
plt.boxplot(data,labels=['Box 1', 'Box 2'])
plt.title('Box Plot of Data')
plt.xlabel('Box')
plt.ylabel('Values')

plt.show()
#A box plot, also known as a box-and-whisker plot, is a statistical visualization tool that provides a summary of the distribution of a dataset. It displays key summary statistics, including the median, quartiles, and potential outliers, in a compact and easily interpretable form.

#Box plots are used for the following purposes:

#Summary of Distribution: Box plots provide a visual summary of the central tendency (median) and the spread (interquartile range) of a dataset. They also show the presence of outliers.

#Comparison of Distributions: Box plots are useful for comparing the distributions of multiple datasets or groups. They allow you to assess differences in central tendency and variability between groups.

#Identification of Outliers: Box plots visually identify potential outliers in the dataset. Outliers are data points that fall significantly outside the whiskers of the plot.
```
