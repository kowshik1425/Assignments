```python
#1
t-test: Used when the population standard deviation is unknown. It estimates the population standard deviation from the sample data. T-tests are generally more robust to violations of normality compared to z-tests, especially for smaller samples.

z-test: Used when the population standard deviation is known and the sample size is large (typically greater than 30).  Z-tests rely on the normal distribution for accurate results.

Here are some example scenarios where you might use each type of test:
Scenario 1: Comparing Mean Heights of Children (t-test)
Imagine you want to compare the average height of boys and girls in a new elementary school. You randomly sample 20 boys and 20 girls and measure their heights.

In this case, you don't necessarily know the population standard deviation for heights of boys and girls in this school.
You would likely use a two-sample t-test to determine if there's a statistically significant difference in the mean heights between boys and girls.
Scenario 2: Checking Blood Pressure after Exercise (z-test)

A study investigates the effect of a new exercise program on blood pressure. They measure the blood pressure of 50 participants before and after the program. They know from previous studies that the standard deviation of blood pressure measurements in the general population is 5 mmHg.
Here, the researchers are interested in the average decrease in blood pressure after the exercise program.
Since the population standard deviation for blood pressure is known (5 mmHg) and the sample size is relatively large (50), they could use a one-sample z-test to see if the observed decrease is statistically significant.
```

```python
#2
One-tailed test:
Also known as a directional or one-sided test.
Hypothesis testing involves testing whether the population parameter is greater than or less than a specified value.
The critical region is located entirely in one tail of the probability distribution.
Used when there is specific directional evidence or hypothesis about the population parameter.
Example:
Testing whether a new drug increases blood pressure (one-tailed test for increase) or decreases blood pressure (one-tailed test for decrease).
Two-tailed test:
Also known as a non-directional or two-sided test.
Hypothesis testing involves testing whether the population parameter is different from a specified value in either direction (greater than or less than).
The critical region is divided between the two tails of the probability distribution.
Used when there is no specific directional evidence or hypothesis about the population parameter.
Example:
Testing whether a new teaching method affects exam scores (two-tailed test for any change).
```

```python
#3
Type I error:
Type I error, also known as a false positive, occurs when the null hypothesis is incorrectly rejected when it is actually true.
It represents the situation where the researcher concludes that there is an effect or difference when there is none in reality.
The probability of making a Type I error is denoted by ùõº, the significance level of the test.
Example scenario:
A medical researcher conducts a clinical trial to test a new drug's effectiveness in reducing blood pressure. If the researcher concludes that the drug is effective (rejects the null hypothesis) when it actually has no effect on blood pressure, this would be a Type I error.
Type II error:
Type II error, also known as a false negative, occurs when the null hypothesis is incorrectly retained when it is actually false.
It represents the situation where the researcher fails to detect an effect or difference when it exists in reality.
The probability of making a Type II error is denoted by ùõΩ
Example scenario:
Continuing with the previous example, if the medical researcher fails to conclude that the new drug is effective (fails to reject the null hypothesis) when it actually does reduce blood pressure, this would be a Type II error.
```

```python
#4
Bayes' theorem is a method for calculating the probability of an event (let's call it Event B) happening given that another event (Event A) has already occurred. It essentially helps us update our initial belief about the probability of Event B after considering new evidence (Event A).

Formula:

P(B | A) = ( P(A | B) * P(B) ) / P(A)

P(B | A): The probability of Event B happening given that Event A has already occurred (what we want to solve for).
P(A | B): The probability of Event A happening given that Event B has already occurred.
P(B): The probability of Event B happening in general (prior probability).
P(A): The probability of Event A happening in general.

Example: Rain and Allergies
Imagine you suffer from allergies and experience itchy eyes (Event B) 20% of the time (P(B) = 0.2). It also tends to rain (Event A) 10% of the time (P(A) = 0.1). However, when it rains, your allergies act up 70% of the time (P(A | B) = 0.7).
Let's say you wake up one morning with itchy eyes.  What's the probability it rained (Event A) given your allergies (Event B)?
Using Bayes' theorem:

P(A | B) = (P(B | A) * P(B)) / P(A)

We are interested in P(A | B), which is the probability of rain given itchy eyes.
P(B | A) = 0.7 (probability of itchy eyes given it rained)
P(B) = 0.2 (general probability of itchy eyes)
P(A) = 0.1 (general probability of rain)

Plugging these values into the formula:

P(A | B) = (0.7 * 0.2) / 0.1  = 0.14 
```

```python
#5
A confidence interval (CI) is a statistical range of values that is estimated to contain the true population parameter with a certain level of confidence. It's a way to express the uncertainty around an estimate based on sample data.

1.Identify the key elements:
Sample mean (xÃÑ): The average of the sample data.
Sample standard deviation (s): The standard deviation of the sample data (a measure of spread).
Sample size (n): The number of observations in the sample.
Confidence level: The percentage of times you expect the confidence interval to capture the true population parameter if you were to repeat the sampling process many times. Common confidence levels are 90%, 95%, and 99%.

2.Find the critical value:This value depends on the chosen confidence level and the degrees of freedom (typically n-1 for a single sample). You can find critical values from a t-distribution table or use software functions.

3.Calculate the margin of error:
Margin of error (ME) = (critical value) * (standard deviation) / ‚àö(sample size)

4.Construct the confidence interval:
Lower bound: sample mean - margin of error
Upper bound: sample mean + margin of error

Example:
Imagine a bakery wants to estimate the average daily number of cupcakes they sell. They randomly sample sales data for 25 days (n=25) and find the following:
Sample mean (xÃÑ) = 50 cupcakes/day
Sample standard deviation (s) = 3 cupcakes/day
Let's calculate a 95% confidence interval for the average daily sales (typical confidence level).

Find the critical value (t-value) for a 95% confidence level with 24 degrees of freedom (n-1). Using a t-distribution table or software, you find the t-value to be approximately 2.060.

Calculate the margin of error:
ME = (2.060) * (3 cupcakes/day) / ‚àö(25 days) ‚âà 2.47 cupcakes/day
Construct the 95% confidence interval:
Lower bound: 50 cupcakes/day - 2.47 cupcakes/day ‚âà 47.53 cupcakes/day
Upper bound: 50 cupcakes/day + 2.47 cupcakes/day ‚âà 52.47 cupcakes/day
```

```python
#6
 Bayes' theorem allows you to calculate the probability of an event (let's call it Event B) happening given that another event (Event A) has already occurred. It essentially helps us update our initial belief about the probability of Event B after considering new evidence (Event A).

Formula:

P(B | A) = ( P(A | B) * P(B) ) / P(A)

P(B | A): The probability of Event B happening given that Event A has already occurred (what we want to solve for).
P(A | B): The probability of Event A happening given that Event B has already occurred.
P(B): The probability of Event B happening in general (prior probability).
P(A): The probability of Event A happening in general.
Sample Problem: Disease Diagnosis

A rare disease affects 0.1% of the population (P(B) = 0.001). A new diagnostic test is being evaluated, and it shows positive results for 90% of people with the disease (P(A | B) = 0.9) but also gives positive results for 5% of healthy people (P(A | not B) = 0.05).
Let's say someone tests positive on the new diagnostic test (Event A). What's the probability they actually have the disease (Event B)?
Solution using Bayes' theorem:
We are interested in P(B | A), the probability of having the disease given a positive test result.
P(A | B) = 0.9 (probability of a positive test given the disease)
P(B) = 0.001 (general probability of having the disease)
We need to find P(A), the general probability of a positive test result (considering both diseased and healthy individuals).

Calculate P(A):
P(A) = (P(A | B) * P(B)) + (P(A | not B) * P(not B))
P(not B) = 1 - P(B) = 0.999 (probability of not having the disease)
P(A) = (0.9 * 0.001) + (0.05 * 0.999) = 0.05045
Now that we have all the values, let's calculate P(B | A):
P(B | A) = (P(A | B) * P(B)) / P(A) = (0.9 * 0.001) / 0.05045 ‚âà 0.0178
Therefore, based on Bayes' theorem, there's only a 1.78% chance that someone with a positive test result actually has the disease. This might seem surprising at first, but it's because the disease is very rare (low P(B)). The test gives positive results for many healthy people (high P(A | not B)), making a positive test less indicative of the disease in this specific scenario.
```

```python
#8
The margin of error (ME) in a confidence interval represents the amount of uncertainty around the estimated population mean. It reflects the range of values within which we are confident the true population mean lies, based on a specific confidence level.

Here's how sample size affects the margin of error:Larger sample size leads to a smaller margin of error: With more data points, the sample mean provides a more accurate estimate of the population mean. Consequently, the margin of error shrinks, resulting in a narrower confidence interval.
Example:
Imagine you want to estimate the average weight of apples in a large orchard.
Scenario A: You randomly pick 10 apples and find an average weight of 150 grams.
Scenario B: You randomly pick 50 apples and find an average weight of 145 grams.
Let's say you calculate a 95% confidence interval for both scenarios. It's likely that the confidence interval in scenario B (with a larger sample size) will be narrower than the one in scenario A (smaller sample size). This means you can be more confident that the true population mean weight of apples falls within the smaller range for scenario B.
A larger sample provides a more precise picture of the population's characteristics. It's like having a clearer snapshot of the entire orchard's apples instead of just a small handful.
With more data points, the random fluctuations that can occur in small samples tend to average out, leading to a more reliable estimate of the central tendency (mean weight in this case).
```
