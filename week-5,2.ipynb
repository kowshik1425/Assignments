```python
#ZeroDivisionError. The error is caught in the except block, and a message is printed indicating the error.

try:
    c=10/0
except ZeroDivisionError as e:
    print("error:",e)
    
#OverflowError:This exception is raised when an arithmetic operation results in a numeric overflow, meaning the result of the operation is too large to be represented using the available data type.

large_number = 10 ** 300  
exponent = 1000

try:
    result = large_number ** exponent
except OverflowError as e:
    print("Error:", e)
```

```python
#The LookupError class is a base class for exceptions that occur when an attempt to access a sequence or a mapping  fails due to invalid indices or keys.
#It's a common parent class for exceptions related to looking up values in data structures. 
my_dict = {"apple": 3, "banana": 5, "orange": 2}

try:
    value = my_dict["grape"]
except KeyError as e:
    print("Error:", e)


my_list = [10, 20, 30, 40, 50]

try:
    value = my_list[10]
except IndexError as e:
    print("Error:", e)
```

```python
#ImportError:ImportError is a base exception class in Python that's raised when there's a problem during the import of a module.
#=>It covers a range of import-related issues, such as inability to locate a module, syntax errors within the module, or other issues related to the module's content.
#=>ImportError can be caught to handle import-related errors in a more controlled manner.

try:
    import pwshills
except ImportError as e:
    print("ImportError:", e)

#ModuleNotFoundError:ModuleNotFoundError is a more specific exception class introduced in Python 3.6 as a subclass of ImportError.
#=>It's raised when Python encounters an issue where the specified module cannot be found during the import process.
#=>ModuleNotFoundError is more focused than the broader ImportError and explicitly indicates that the main problem is the inability to locate the specified module.

try:
    import pwskills
except ModuleNotFoundError as e:
    print("ModuleNotFoundError:", e)
```
