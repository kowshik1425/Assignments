```python
#1
from scipy.stats import norm
sample_mean = 50
sample_std_dev = 5
confidence_level = 0.95
sample_size = 100 
standard_error = sample_std_dev / (sample_size ** 0.5)
critical_value = norm.ppf((1 + confidence_level) / 2)
margin_of_error = critical_value * standard_error
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error
print("95% Confidence Interval: ({:.2f}, {:.2f})".format(lower_bound, upper_bound))
```

```python
#4
import numpy as np
from scipy.stats import norm
n = 500
x = 60
p_hat = x / n
standard_error = np.sqrt(p_hat * (1 - p_hat) / n)
critical_value = norm.ppf(0.975)  # 0.975 because it's a two-tailed test for 95% CI
margin_of_error = critical_value * standard_error
lower_bound = p_hat - margin_of_error
upper_bound = p_hat + margin_of_error

print("95% Confidence Interval for the true proportion of smokers:", (lower_bound, upper_bound))
```

```python
#8
To conduct a chi-square goodness of fit test to determine if the observed frequencies match the expected frequencies of a fair coin, we need to follow these steps:

Define the null hypothesis and the alternative hypothesis.
Determine the expected frequencies under the assumption of a fair coin.
Calculate the chi-square statistic based on the observed and expected frequencies.
Determine the degrees of freedom.
Find the critical chi-square value for the chosen significance level.
Compare the calculated chi-square statistic with the critical chi-square value.
Make a decision to either reject or fail to reject the null hypothesis.

import numpy as np
from scipy.stats import chi2
total_flips = 100
expected_tails = total_flips / 2
expected_heads = total_flips / 2
expected_frequencies = [expected_heads, expected_tails]
observed_frequencies = [55, 45]  
chi2_stat = np.sum((np.array(observed_frequencies) - np.array(expected_frequencies))**2 / np.array(expected_frequencies))
degrees_of_freedom = len(observed_frequencies) - 1
critical_value = chi2.ppf(0.95, degrees_of_freedom)  # 0.05 significance level corresponds to 0.95 confidence level
if chi2_stat > critical_value:
    print("Reject the null hypothesis: The observed frequencies do not match the expected frequencies (p < 0.05).")
else:
    print("Fail to reject the null hypothesis: The observed frequencies match the expected frequencies (p >= 0.05).")
```

```python
#11
import numpy as np
from scipy.stats import t

# Sample statistics
sample_mean = 72
sample_std_dev = 10
population_mean = 70
sample_size = 30

# Step 1: Set up the null and alternative hypotheses
# Null hypothesis: Population mean is equal to 70 ( = 70)
# Alternative hypothesis: Population mean is not equal to 70 (  70)

# Step 2: Calculate the t-statistic
t_stat = (sample_mean - population_mean) / (sample_std_dev / np.sqrt(sample_size))

# Step 3: Determine the degrees of freedom
degrees_of_freedom = sample_size - 1

# Step 4: Find the critical t-value for a significance level of 0.05 and degrees of freedom
critical_value = t.ppf(0.975, degrees_of_freedom)  # 0.025 on each tail for a two-tailed test

# Step 5: Compare the calculated t-statistic with the critical t-value
if np.abs(t_stat) > critical_value:
    print("Reject the null hypothesis: Population mean is significantly different from 70 (p < 0.05).")
else:
    print("Fail to reject the null hypothesis: Population mean is not significantly different from 70 (p >= 0.05).")
```
