```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]
p = figure( x_axis_label='X-Axis', y_axis_label='Y-Axis')
p.line(x, y, line_width=2, legend_label="Line")
p.legend.location = "top_left"
p.legend.title = "Legend"
output_notebook()
show(p)
from bokeh.io import output_file, save

output_file("example_plot.html")
save(p)
```

```python
# glyphs are visual markers or shapes that represent data points in a plot. Glyphs are used to visually encode and display data. Bokeh provides a variety of glyph types, such as lines, circles, squares, bars, and more, that you can add to a plot to create different types of visualizations. You can customize the appearance of glyphs to convey information effectively.
from bokeh.plotting import figure, show, output_notebook
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]
p = figure(title="Glyph Example", x_axis_label='X-Axis', y_axis_label='Y-Axis')
p.circle(x, y, size=10, color='blue', alpha=0.5, legend_label='Circles')

output_notebook()
show(p)
```

```python
Customizing the Plot Title:To customize the title of your Bokeh plot, you can set the title attribute of the figure 
from bokeh.plotting import figure, show
p = figure(title="Customized Title")

Customizing the Axes Labels:You can set the labels for the x-axis and y-axis using the x_axis_label and y_axis_label attributes.
p = figure(x_axis_label="Custom X-Axis Label", y_axis_label="Custom Y-Axis Label")

Customizing the Axes Ticks and Grid Lines:You can customize the appearance of the axis ticks and grid lines using attributes like xaxis, yaxis, xgrid, and ygrid. For example, to change the color of the x-axis.
p.xaxis.major_label_text_color = "red"

Customizing Plot Dimensions:To customize the dimensions of the plot, you can use the width and height attributes when creating the figure object.
p = figure(width=800, height=400)
```

```python
#A Bokeh server is a component of the Bokeh library that allows you to create and deploy interactive web applications with real-time updates. It enables you to build dynamic and interactive data visualizations and dashboards that can be accessed through a web browser. Bokeh servers are particularly useful when you need to create applications that involve user interactions, live data streaming, or real-time data updates.
#Create a Bokeh Application:To create a Bokeh server application, you'll need to define a Python script that creates the interactive plot and specifies how it should react to user interactions or data updates. You can use Bokeh's curdoc() function to access the current document and make updates to the plot.

# Import necessary modules
from bokeh.plotting import figure, curdoc
from bokeh.models import ColumnDataSource
from bokeh.io import output_notebook, push_notebook
from bokeh.layouts import column
import numpy as np
x = np.linspace(0, 10, 100)
y = np.sin(x)
source = ColumnDataSource(data={'x': x, 'y': y})
plot = figure(title="Real-time Line Plot")
line = plot.line('x', 'y', source=source)
def update():
    new_y = np.sin(x + 0.1)
    source.data = {'x': x, 'y': new_y}
curdoc().add_periodic_callback(update, 100)
curdoc().title = "Real-time Plot Example"
curdoc().add_root(plot)
```

```python
#You can embed a Bokeh plot into a web page or dashboard using Flask or Django by generating the Bokeh plot as an HTML document and then including that HTML document within your Flask or Django template. 

from flask import Flask, render_template
from bokeh.plotting import figure, show
from bokeh.embed import components

app = Flask(__name__)

@app.route('/')
def index():
    p = figure(title="Bokeh Plot")
    p.circle([1, 2, 3], [4, 5, 6])
    script, div = components(p)
    return render_template('index.html', script=script, div=div)

if __name__ == '__main__':
    app.run()

```
